```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
 
scaled_x_train = sc.fit_transform(x_train)
scaled_x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
```

```python
lr_model
```

```python
lr_model.fit(scaled_x_train,y_train)
```

```python
lr_pred = lr_model.predict(scaled_x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
accuracy_score(y_test,lr_pred)
```

```python
confusion_matrix(y_test,lr_pred)
```

```python
classification_report(y_test,lr_pred)
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model
```

```python
svc_model.fit(x_train, y_train)
y_pred_svc = svc_model.predict(x_test)
```

```python
print('accuracy_score of SVC : ', accuracy_score(y_test, y_pred_svc))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model
```

```python
knn_params = {
    'n_neighbors' : [x for x in range(10, 30, 2)],
    'metric' : ['euclidean', 'manhattan']
}
```

```python
from sklearn.model_selection import GridSearchCV
knn_gcv = GridSearchCV(knn_model, knn_params, cv=5)
knn_gcv.fit(x_train, y_train)
print(knn_gcv.best_params_)
```

```python
knc = KNeighborsClassifier(n_neighbors=10, metric='manhattan')
knc.fit(x_train, y_train)
y_pred_knc = knc.predict(x_test)
```

```python
print('accuracy_score of KNeighborsClassifier : ', accuracy_score(y_test, y_pred_knc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
rfc_model
```

```python
rfc_params = {
    'n_estimators' : [x for x in range(10, 100, 10)],
    'criterion' : ['gini', 'entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
rfc_gcv = GridSearchCV(rfc_model, rfc_params, cv=5)
rfc_gcv.fit(x_train, y_train)
print(rfc_gcv.best_params_)
```

```python
rfc = RandomForestClassifier(n_estimators=90, criterion='entropy')
rfc.fit(x_train, y_train)
y_pred_rfc = rfc.predict(x_test)
```

```python
print('accuracy_score of RandomForestClassifier : ', accuracy_score(y_test, y_pred_rfc))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc_model = DecisionTreeClassifier()
dtc_model
```

```python
dtc_params = {
    'criterion' : ['gini', 'entropy'],
    'splitter' : ['best']
}
```

```python
dtc_gcv = GridSearchCV(dtc_model, dtc_params, cv=5)
dtc_gcv.fit(x_train, y_train)
print(dtc_gcv.best_params_)
```

```python
dtc = DecisionTreeClassifier(criterion='entropy', splitter='best')
dtc.fit(x_train, y_train)
y_pred_dtc = dtc.predict(x_test)
```

```python
print('accuracy_score of DecisionTreeClassifier : ', accuracy_score(y_test, y_pred_dtc))
```

```python
print(classification_report(rfc.predict(x_test), y_test))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
y_pred = rfc.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rfc.classes_)
disp.plot(cmap='Blues')
plt.show()
```
